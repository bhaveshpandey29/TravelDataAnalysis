## Clustering

```python
df_cluster = df.copy()
```

```python
from sklearn import preprocessing
```

```python
x = df_cluster.iloc[:,14:16]
x
```

```python
x_scaled = preprocessing.scale(x)
```

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 3
```

```python
kmeans = KMeans(n_clusters = n_clusters, init='k-means++', n_init = 10 ,max_iter=300,tol=0.0001,  random_state= 111  , algorithm='elkan')
```

```python
kmeans.fit(x_scaled)
```

```python
identified_clusters = kmeans.fit_predict(x_scaled)
identified_clusters
```

```python
data_with_clusters = df_cluster.copy()
data_with_clusters['cluster'] = identified_clusters
centroids = pd.DataFrame(kmeans.cluster_centers_,columns=["ACTIVE_TRAVEL_TIME","CAR_TIME"])
#data_with_clusters
```

```python
#from sklearn.preprocessing import LabelEncoder
#encode = LabelEncoder()
#encode.fit(data_with_clusters['STATE'])
#data_with_clusters['STATE'] = encode.transform(data_with_clusters['STATE'])
```

```python
centroid = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(15,12))

scatter = plt.scatter(data_with_clusters['ACTIVE_TRAVEL_TIME'], data_with_clusters['CAR_TIME'] ,cmap="rainbow",c= data_with_clusters['cluster'],alpha=0.8, s=13,edgecolors='none')
plt.xlabel('Active Travel Time in Minute'), plt.ylabel('Car Travel Time in Minute') 
classes = ['Group 1', 'Group 2', 'Group 3', 'Group 4','Group 5','Group 6']
rangex = list(range(0,550,50))
rangey = list(range(0,850,50))
plt.xticks(rangex)
plt.yticks(rangey)

#plt.xlim(l100)
#plt.legend(*scatter.legend_elements())
plt.legend(handles = scatter.legend_elements()[0],labels = classes)
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
#exlain variance

pca = PCA().fit(df2)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca = PCA(n_components = 2)
pca.fit_transform(df2)
print(pca.explained_variance_ratio_)
```

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
pca_components=pd.DataFrame(columns=['pca'])
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pcs = pca.fit_transform(df2)
df_new = pd.DataFrame(data=pcs, columns={'PC1','PC2'})
df_new['target'] = target_cluster
df_new.head()
```
