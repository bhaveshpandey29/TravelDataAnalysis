```python
n_clusters = 3
```

```python
kmeans = KMeans(n_clusters = n_clusters, init='k-means++', n_init = 10 ,max_iter=300,tol=0.0001,  random_state= 111  , algorithm='elkan')
```

```python
kmeans.fit(x_scaled)
```

```python
identified_clusters = kmeans.fit_predict(x_scaled)
identified_clusters
```

```python
data_with_clusters = df_cluster.copy()
data_with_clusters['cluster'] = identified_clusters
centroids = pd.DataFrame(kmeans.cluster_centers_,columns=["ACTIVE_TRAVEL_TIME","CAR_TIME"])
#data_with_clusters
```

```python
from sklearn.decomposition import PCA
```

```python
#exlain variance

pca = PCA().fit(df2)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca = PCA(n_components = 2)
pca.fit_transform(df2)
print(pca.explained_variance_ratio_)
```

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
pca_components=pd.DataFrame(columns=['pca'])
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pcs = pca.fit_transform(df2)
df_new = pd.DataFrame(data=pcs, columns={'PC1','PC2'})
df_new['target'] = target_cluster
df_new.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0,1,2]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_new['target'] == target_cluster
    ax.scatter(df_new.loc[indicesToKeep, 'PC1']
               , df_new.loc[indicesToKeep, 'PC2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
